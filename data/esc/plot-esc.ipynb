```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------------------------
# 1)  Load embeddings-metadata file
# -------------------------------------------------
csv_path = "/Users/sebasmos/Documents/DATASETS/data_VE/ESC-50-master/VE_soundscapes/efficientnet_1536/esc-50.csv"
df = pd.read_csv(csv_path)
print(f"Loaded {len(df):,} rows")

# -------------------------------------------------
# 2)  Bar-plot: category / class distribution
# -------------------------------------------------
plt.figure(figsize=(10,4))
sns.countplot(data=df, x="category", order=df["category"].value_counts().index,
              palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("ESC-50  examples per 5 super-classes")
plt.ylabel("Count")
plt.xlabel("Category")
plt.tight_layout()
plt.show()

# -------------------------------------------------
# 3)  Bar-plot: fold distribution
# -------------------------------------------------
plt.figure(figsize=(6,3))
sns.countplot(data=df, x="folder", palette="crest")
plt.title("ESC-50  fold distribution")
plt.ylabel("Count")
plt.xlabel("Fold")
plt.tight_layout()
plt.show()
```

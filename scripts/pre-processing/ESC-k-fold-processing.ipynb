```python
import os
import pandas as pd
import shutil


# Create the output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Read the CSV file
df = pd.read_csv(csv_path)

# Get unique folds from the CSV
unique_folds = df['fold'].unique()

# Create the output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Extract unique class names from folder names in data_path and map them to integers
class_folders = sorted([d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))])
class_to_int = {class_name: idx for idx, class_name in enumerate(class_folders)}

print("Class Mapping:", class_to_int)  # Debugging line to check mapping

# Get unique folds from the CSV
unique_folds = df['fold'].unique()

# Process each fold
for fold in unique_folds:
    fold_dir = os.path.join(output_path, f'fold_{fold}')
    os.makedirs(fold_dir, exist_ok=True)

    # Create CSV file to store image names and corresponding class labels
    fold_csv_path = os.path.join(output_path, f'fold_{fold}.csv')
    image_class_data = []

    # Get filenames that match the current fold
    matching_files = df[df['fold'] == fold]['filename'].str.replace('.wav', '.png').tolist()

    print(f"\nProcessing fold {fold} with {len(matching_files)} files.")  # Debugging line

    for class_name, class_label in class_to_int.items():  # Iterate through class folders
        class_path = os.path.join(data_path, class_name)
        if os.path.isdir(class_path):
            existing_files = os.listdir(class_path)  # Get all files in class folder

            for filename in matching_files:
                if filename in existing_files:
                    source_path = os.path.join(class_path, filename)
                    dest_path = os.path.join(fold_dir, filename)

                    print(f"Copying {source_path} â†’ {dest_path}")  # Debugging line
                    shutil.copy(source_path, fold_dir)  # Copy file to fold directory

                    # Save image name and class label
                    image_class_data.append((filename, class_label))

    # Save fold CSV with image names and class labels
    fold_df = pd.DataFrame(image_class_data, columns=["image_name", "class"])
    fold_df.to_csv(fold_csv_path, index=False)
    print(f"Saved {fold_csv_path} with {len(fold_df)} entries.")

print("\nFolders and class mapping files created successfully.")
```
